```python
import matplotlib.pyplot as plt

patch_sizes = [3, 5, 7, 16]
success_rates = [50.30, 52.88, 55.66, 77.38]

plt.figure(figsize=(5, 4))
plt.plot(patch_sizes, success_rates, marker='o', color='r', label='Attack Success Rate')

# plt.title('Fig 1. Attack Success Rate\nfor Different Patch Sizes')
plt.xlabel('Patch Size (pixels)')
plt.ylabel('Attack Success Rate (%)')

plt.xlim(2, 17)
plt.ylim(40, 80)

# plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

patch_sizes = [3, 5, 7, 16]
patch_areas = [3**2, 5**2, 7**2, 16**2]
image_area = 32 * 32
patch_size_percentages = [area / image_area * 100 for area in patch_areas]
success_rates = [50.30, 52.88, 55.66, 77.38]

plt.figure(figsize=(5, 4))
plt.plot(patch_size_percentages, success_rates, marker='o', color='b', label='Attack Success Rate')

# plt.title('Fig 2. Attack Success Rate by Patch Size\n(% of Image Size)', pad=10)
plt.xlabel('Patch Size (% of Image Size)')
plt.ylabel('Attack Success Rate (%)')
plt.ylim(40, 80)

# plt.grid(True)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

patch_sizes = ['3x3', '5x5']
num_patches = [1, 2, 3, 4, 5]

attack_success_3x3 = [64.54, 69.10, 71.83, 71.79, 72.73]
attack_success_5x5 = [72.18, 76.87, 77.46, 78.96, 78.17]

plt.figure(figsize=(5, 4))
plt.plot(num_patches, attack_success_3x3, label='3x3 Patch', marker='o', linestyle='-', color='b')
plt.plot(num_patches, attack_success_5x5, label='5x5 Patch', marker='s', linestyle='--', color='r')

# plt.title('Fig 3. Untargeted Attack Success Rate\nfor Varying Patch Sizes and Counts',
#           fontsize=14, pad=10)
plt.xlabel('Number of Patches', fontsize=12)
plt.ylabel('Attack Success Rate (%)', fontsize=12)
plt.xticks(np.arange(1, 6))
plt.yticks(np.arange(60, 81, 5))
plt.legend()

# plt.grid(True)
plt.tight_layout()
plt.show()
```
